## 🛠️ 环境设置与安装 (Setup & Installation)

本项目在 **Python 3.12** 环境下开发和测试。

1.  **克隆仓库**
    ```bash
    git clone [你的仓库HTTPS链接]
    cd logistic_regression_project
    ```

2.  **创建并激活 Conda 虚拟环境**
    ```bash
    # 创建环境
    conda create --name logistic_env python=3.12

    # 激活环境
    conda activate logistic_env
    ```

3.  **安装依赖**
    项目所需的所有依赖库都记录在 `requirements.txt` 文件中。请运行以下命令进行安装：
    ```bash
    pip install -r requirements.txt
    ```
